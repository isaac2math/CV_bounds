```python
%reset -f # clear the memory before each compilation

import numpy as np
from scipy.sparse.linalg import eigsh
import matplotlib.pyplot as plt

from sklearn.linear_model import Lasso, LassoCV
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
```

```python
##################
# data generator #
##################

######### verified by test environment ############


def data_gen(sample_size, n_dim, n_info, cov_noise):
    
    a = np.ones((n_dim, n_dim)) * 0.5; A = np.eye(n_dim)*0.5

    cov_x = a + A; mean_x = np.zeros(n_dim)

    X = np.random.multivariate_normal(mean_x, cov_x, sample_size)

    beta_info = np.arange(1,n_info + 1)
    beta = np.concatenate((beta_info, np.zeros(n_dim-n_info)), axis = 0)

    noise = np.random.normal(0, cov_noise, sample_size); 
    noise.shape = (sample_size, 1); beta.shape = (1, n_dim) 

    Y = np.inner(X,beta) + noise 

    return X, Y
```

```python
##################################
# Rademacher variables generator #
##################################

######### verified by test environment ############

def rade_generator( sample_size ):
    
    ans = np.random.randint(2, size=sample_size)
    rade = (ans - 0.5) * 2
    
    return rade
```

```python
##############################################
# estimating empirical rademacher complexity #
##############################################

######### verified by test environment ############


def emp_rc(sample_size, n_iter, reg, n_dim, n_info, cov_noise, variable_mark):
    
    X, Y = data_gen(sample_size, n_dim, n_info, cov_noise)
    X_input = X[:, variable_mark]
    
    cond = np.zeros([n_iter,1])
    
    for i in range(n_iter):
        
        rade = rade_generator(sample_size)
        reg.fit(X_input, rade) ; rade_pre = reg.predict(X_input)
        rade_pre = np.matrix(rade_pre); cond[i,0] = np.dot(rade_pre, rade)/ sample_size
    
    emp_rc = np.mean(cond) * 2 
   
    return emp_rc, cond
```

```python
####################################
# estimating rademacher complexity #
####################################

######### verified by test environment ############
def est_rc(sample_size, n_iter, reg, n_dim, n_info, cov_noise, variable_mark):
      
    cond_rc = np.zeros([n_iter])
    
    for i in range(n_iter):
        
        emp_rc_in, _ = emp_rc(sample_size, n_iter, reg, n_dim, n_info, cov_noise, variable_mark)

        cond_rc[i] = emp_rc_in

    est_rc = np.mean(cond_rc)
    
    return est_rc, cond_rc
```

```python
######################################################
# establishing error-stability bound for classifiers #
######################################################

######### verified by test environment ############

def rc_bound(n, M, varpi, K, rc_1, rc_2, ave_in_sample_error):
    
    ub = ave_in_sample_error + rc_1 + rc_2 + 2 * M * np.sqrt( np.log( 1 / varpi ) / ( n / K ) )
    
    return ub  
```

```python
#########################################################
# establishing the beta-stability bound for classifiers #
#########################################################

def b_bound(n, M, varpi, K, rc_1, rc_2, training_error, test_error, exog):
    
    ub = training_error - test_error + exog + rc_1 + rc_2 + 2 * M * np.sqrt( np.log( 1 / varpi ) / ( n / K ) )
    
    return ub  
```

```python
##################################################
# compute the mean square error of the estimator #
##################################################

######### verified by test environment ############


def costs(X_train, Y_train, clf):
    
    Y_pred_train = clf.predict(X_train); Y_pred_train.shape = Y_train.shape
    loss_train = mean_squared_error(Y_train, Y_pred_train)
    res = Y_train - Y_pred_train; res_var = np.var(res)
        
    return loss_train, res_var, res
```

```python
###########################################################
# the function to estimate the upper bounds for the lasso #
###########################################################

def error_bounds(X_train_1, Y_train_1, X_train_2, Y_train_2, 
              test_size, varpi, n_iter, n_val, n_dim, n_info, cov_noise, start, end, step):
    
    ##set the placeholders 
    alpha = np.arange(start, end, step)          # the seq of penalty parameters
    error_matrix = np.zeros((n_val, len(alpha))) # the n_val out-of-sample errors for each model
    upper_array  = np.zeros(len(alpha))          # the seq of average test errors for all models
    train_array  = np.zeros(len(alpha))          # the seq of average trainig errors for all models
    test_array   = np.zeros(len(alpha))          # the seq of 90 percentiles for all models
    perc_array   = np.zeros(len(alpha))
    
    for l in range(len(alpha)):
        
        #given the value of alpha
        
        #training classifiers
        clf  = Lasso(alpha = alpha[l])
        clf1 = Lasso(alpha = alpha[l]); clf1.fit(X_train_1, Y_train_1)
        clf2 = Lasso(alpha = alpha[l]); clf2.fit(X_train_2, Y_train_2)
        clf3 = Lasso(alpha = alpha[l]); clf2.fit(X_train_2, Y_train_2)
    
        #compute training errors
        loss_train_1, res_var_1 = costs(X_train_1, Y_train_1, clf1)
        loss_train_2, res_var_2 = costs(X_train_2, Y_train_2, clf2)
        train_array[l] = (loss_train_1 + loss_train_2)/2
        
        #compute test errors
        loss_test_1, res_var_3 = costs(X_train_2, Y_train_2, clf1)
        loss_test_2, res_var_4 = costs(X_train_1, Y_train_1, clf2)
        test_array[l] = (loss_test_1 + loss_test_2)/2
        
        res_var = max(res_var_3, res_var_4) ; M = res_var * 1.4
        

        #container of the validation errors
        error_val = np.zeros(n_val); 

        #the loop to compute the out-of-sample errors
        for i in range(n_val):

            X_val_1, Y_val_1 = data_gen(test_size, n_dim, n_info, cov_noise)
            X_val_2, Y_val_2 = data_gen(test_size, n_dim, n_info, cov_noise)

            Y_pred_val_1 = clf1.predict(X_val_1) ; loss_test_val_1 = mean_squared_error(Y_val_1 , Y_pred_val_1)
            Y_pred_val_2 = clf2.predict(X_val_2) ; loss_test_val_2 = mean_squared_error(Y_val_2 , Y_pred_val_2)

            error_val[i] = (loss_test_val_1 + loss_test_val_2)/2
        
        error_matrix[:,l] = error_val
        
        sorted_error = np.sort(error_val); perc_array[l] = sorted_error[-1*round(n_val*0.1)]

        # estimating the RC_test and RC_train
        est_rc_train, cond_rc_SVC_train = est_rc(test_size, n_iter, clf, n_dim, n_info, cov_noise)
        est_rc_test, cond_rc_SVC_test = est_rc(test_size, n_iter, clf, n_dim, n_info, cov_noise)

        #establish the upper bounds
        upper_array[l] = rc_bound( 2 * test_size, M, varpi, K, est_rc_train, est_rc_test, train_array[l])
        
        print("Model", int(l+1), " computed out of ", len(alpha))
        
    
    return error_matrix, upper_array, train_array, test_array, perc_array
```

```python
###################################################
######### the relaxed-lasso package ###############
###################################################

######### verified by test environment ############

def r_lasso(X_train_1, Y_train_1, X_train_2, Y_train_2, alpha):
    
    ##set the placeholders 
    train_array  = np.zeros(len(alpha))          # the seq of average trainig errors for all models
    test_array   = np.zeros(len(alpha))          # the seq of average test errors for all models
    

    ##select the optimal value of the penalty parameter for the lasso
    for l in range(len(alpha)):
        
        clf_1  = Lasso(alpha = alpha[l]); clf_1.fit(X_train_1, Y_train_1) #the clf for 1st round
        clf_2  = Lasso(alpha = alpha[l]); clf_2.fit(X_train_2, Y_train_2)   #the clf for 2nd round
        
        #compute training errors
        loss_train_1,_ ,_ = costs(X_train_1, Y_train_1, clf1)
        loss_train_2,_ ,_ = costs(X_train_2, Y_train_2, clf2)
        train_array[l] = (loss_train_1 + loss_train_2)/2
                
        #compute test errors
        loss_test_1,_ ,_ = costs(X_train_2, Y_train_2, clf1)
        loss_test_2,_ ,_ = costs(X_train_1, Y_train_1, clf2)
        test_array[l] = (loss_test_1 + loss_test_2)/2
    
    min_loc = np.where(test_array == min(test_array)); opt_alpha = alpha[min_loc]

    ##concatenate the train and test data, retrain the lasso on the whole data
    
    # concatenate training and test samples
    X = np.concatenate((X_train_1, X_train_2), axis = 0)
    Y = np.concatenate((Y_train_1, Y_train_2), axis = 0)

    #restrain the lasso with optimal alpha on (X,Y)
    #and identify the selected variables
    clf_overall  = Lasso(alpha = opt_alpha); clf_overall.fit(X, Y) 
    feature_index = clf_overall.coef_; feature_index[feature_index > 0] = 1
    variable_mark = (feature_index == 1)
    
    #train an OLS with the selected variable on two training sets
    clf_out_1 = LinearRegression(); clf_out_1.fit(X_train_1[:,variable_mark],Y_train_1)
    clf_out_2 = LinearRegression(); clf_out_2.fit(X_train_2[:,variable_mark],Y_train_2)
    norm_ave =  0.5 * np.linalg.norm(clf_out_1.coef_, 2) + 0.5 * np.linalg.norm(clf_out_2.coef_, 2)
    
    #compute the test error of the OLS
    loss_test_out_1, res_out_1, res_te_1 = costs(X_train_2[:,variable_mark], Y_train_2, clf_out_1)
    loss_test_out_2, res_out_2, res_te_2 = costs(X_train_1[:,variable_mark], Y_train_1, clf_out_2)
    test_out = (loss_test_out_1 + loss_test_out_2)/2
    res_test = max(res_out_1, res_out_2)
    
    #compute the training error of the OLS
    loss_train_out_1, res_out_3,_ = costs(X_train_2[:,variable_mark], Y_train_2, clf_out_2)
    loss_train_out_2, res_out_4,_ = costs(X_train_1[:,variable_mark], Y_train_1, clf_out_1)
    train_out = (loss_train_out_1 + loss_train_out_2)/2
    res_train = max(res_out_3, res_out_4)
    
    return res_train, res_test, train_out, test_out, variable_mark, res_te_1, res_te_2, norm_ave
```

```python

import numpy as np
import matplotlib.pyplot as plt
from scipy.sparse.linalg import eigsh


from sklearn.linear_model import Lasso, LassoCV
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression

################################################################
# the beta_stability bounds for the lasso, based on the rlasso #
################################################################

def beta_bounds(X_train_1, Y_train_1, X_train_2, Y_train_2, 
              test_size, varpi, n_iter, n_val, n_dim, n_info, cov_noise, start, end, step):
    
    ##set the placeholders 
    alpha = np.arange(start, end, step)    #the seq of penalty parameters
    norm_val = np.zeros(n_val)             #container of the L_1 norm of the lasso in simulations
    clf_trial = LassoCV(cv = 2)

    
    #train the rlasso and select correct variables
    res_train, res_test, train_out, test_out, variable_mark, res_te_1, res_te_2, norm_ave = r_lasso(X_train_1, Y_train_1, X_train_2, Y_train_2, alpha)
    
    #compute M
    res_var = max(res_te_1, res_te_2) ; M = res_var * 1.4

    #compute the RC for the rlasso
    RC_1 = est_rc(sample_size, n_iter, reg, n_dim, n_info, cov_noise, variable_mark)
    RC_2 = est_rc(sample_size, n_iter, reg, n_dim, n_info, cov_noise, variable_mark)
    
    #compute the exog for the rlasso
    
    exog_1 = np.linalg.norm(np.matmul(res_te_1.T, X_train_2), 2) * norm_ave
    exog_2 = np.linalg.norm(np.matmul(res_te_2.T, X_train_1), 2) * norm_ave
    exog_ave = np.mean(exog_1, exog_2)

    #compute the sparse eigenvalue of the Gram matrix
    
    Gram_1 = np.matmul(X_test_1.T, X_test_1); Gram_2 = np.matmul(X_test_2.T, X_test_2) 
    sp_eigen_1 = eigsh(Gram_1.T); sp_eigen_2 = eigsh(Gram_2.T)
    min_sp_eigen = np.mean(min(sp_eigen_2), min(sp_eigen_1))

    #establish the upper bounds
    pred_upper_bound = b_bound( 2 * test_size, M, varpi, K, RC_1, RC_2, train_out, test_out, exog_ave)

    beta_upper_bound = pred_upper_bound / min_sp_eigen
    
    #the loop to compute the out-of-sample errors
    for i in range(n_val):

        X_val, Y_val = data_gen(test_size, n_dim, n_info, cov_noise)

        clf_trial.fit(X_val, Y_val); reg_c = clf_trial.coef_

        norm_val[i] = np.linalg.norm(reg_c, 1)

        print("iter", int(i+1), " computed out of ", len(n_val))

    sorted_norm = np.sort(norm_val); perc_norm = sorted_norm[-1*round(n_val*0.1)]
    
    return norm_val, perc_norm, upper_bound, pred_upper_bound, beta_upper_bound
```

```python
##################
# result plotter #
##################

######### verified by test environment ############


def error_plot(alpha, error_matrix, upper_array, train_array, test_array, perc_array,name):
    
    #set the tick of the x axis
    index_plot = np.arange(1, len(alpha)+1); #my_xticks = list(map(str,alpha))
    my_xticks = ['0.1', '0.15', '0.2', '0.25', '0.3', '0.35', '0.4', '0.45']
    
    #set the figure
    f = plt.figure()
    
    #multiple boxplot
    plt.boxplot(error_matrix, widths = 0.25)

    #plot the average training error, average test error, 90% percentile and 90% upper bound
    plt.plot(index_plot, upper_array, color = 'g', label='at-least 90% upper bound of the out-of-sample error')
    plt.plot(index_plot, train_array, color = 'b', label='the average training error')
    plt.plot(index_plot, test_array,  color = 'r', label='the average test error')
    plt.plot(index_plot, perc_array,  color = 'k', label='90% percentile')

    #set the legend, label, grid and tick
    legend = plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0., shadow=True)
    frame = legend.get_frame()

    plt.xticks(index_plot, my_xticks)
    plt.grid(axis='y', linestyle='-')
    #plt.title('the out-of-sample error of the lasso')
    plt.xlabel('value of $\lambda$')
    plt.ylabel('value of the error')

    plt.show()
    
    # save the plot
    f.savefig(name, bbox_inches='tight')
```

```python
## average test error bounds for the lasso
## total sample size is 200
## true model : y = x1 + 2 * x2 + 3 * x3 + 4 * x4 + 5 * x5 
                # + 0 * x6 + 0 * x7 + 0 * x8 + 0 * x9 + 0 * x10 + e

n2 = 100; test_size = n2; np.random.seed(5) 

# generate data

varpi = 0.1; K = 2

n_iter = 128; n_val = 10000; n_dim = 100; n_info = 5; cov_noise = 1

start = 0.1; end = 0.5; step = 0.05; alpha = np.arange(start, end, step)


X12, Y12 = data_gen(n2, n_dim, n_info, cov_noise)

X22, Y22 = data_gen(n2, n_dim, n_info, cov_noise)


# training the model
error_matrix_2, upper_array_2, train_array_2, test_array_2, perc_array_2 = error_bounds(X12, Y12, X22, Y22, test_size, varpi, 
                                                                              n_iter, n_val, n_dim, n_info, cov_noise, start, end, step)

# plot the figure
error_plot(alpha, error_matrix_2, upper_array_2, train_array_2, test_array_2, perc_array_2, 
           "error_bound_plot_200.pdf")
```
